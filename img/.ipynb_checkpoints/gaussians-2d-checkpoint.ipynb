```python
import arviz as az
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

az.style.use("arviz-darkgrid")
plt.rcParams.update({"figure.dpi": 300})
```

```python
X, Y = np.linspace(-5, 5, 50), np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
XY = np.dstack((X, Y))

means = [0, 0]
cov1 = np.diag(np.array([4, 1]))
cov2 = np.diag(np.array([7, 7]))

rv1 = scipy.stats.multivariate_normal(means, cov1)
rv2 = scipy.stats.multivariate_normal(means, cov2)
Z1 = rv1.pdf(XY)
Z2 = rv2.pdf(XY)

fig = plt.figure(figsize=plt.figaspect(0.5))
fig.suptitle("Gaussian distributions centered at $(\mu_x, \mu_y)=(0, 0)$")
ax1 = fig.add_subplot(1, 2, 1, projection="3d")
ax1.plot_surface(X, Y, Z1, cmap="viridis")
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_zticks([])
ax1.set_title("($\sigma_x, \sigma_y) = (4, 1)$")
ax2 = fig.add_subplot(1, 2, 2, projection="3d")
ax2.plot_surface(X, Y, Z2, cmap="viridis")
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_zticks([])
ax2.set_title("$(\sigma_x, \sigma_y) = (7, 7)$")

plt.show()
```

```python
fig.savefig("gaussians-2d.png")
```
